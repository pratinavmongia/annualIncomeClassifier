```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Drop rows with '?' in any column
df_cleaned = df1[~df1.astype(str).apply(lambda x: x.str.contains(r'\?')).any(axis=1)]
df_cleaned.head()
```

Training and testing data Split And Transforming the Classification
Columns using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
pd.__version__
```

```python
pd.options.mode.copy_on_write = True 
```

```python
lb = LabelEncoder()
df_cleaned['workclass'] = lb.fit_transform(df_cleaned['workclass'])
df_cleaned['education'] = lb.fit_transform(df_cleaned['education'])
df_cleaned['marital.status'] = lb.fit_transform(df_cleaned['marital.status'])
df_cleaned['occupation'] = lb.fit_transform(df_cleaned['occupation'])
df_cleaned['relationship'] = lb.fit_transform(df_cleaned['relationship'])
df_cleaned['race'] = lb.fit_transform(df_cleaned['race'])
df_cleaned['sex'] = lb.fit_transform(df_cleaned['sex'])
df_cleaned['native.country'] = lb.fit_transform(df_cleaned['native.country'])
```

```python
x=df_cleaned[['age', 'workclass', 'fnlwgt', 'education', 'education.num',
       'marital.status', 'occupation', 'relationship', 'race', 'sex',
       'capital.gain', 'capital.loss', 'hours.per.week', 'native.country']]
y=df_cleaned['income']
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Decision Tree Classifier


```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def gen_metrics(ytest,ypred):
    cm = confusion_matrix(ytest,ypred)
    print('confusion matrix\n',cm)
    print('Classification report\n',classification_report(ytest,ypred))
    print('Acc Score', accuracy_score(ytest,ypred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
m1 = DecisionTreeClassifier()
m1.fit(x_train,y_train)
```

```python
#Accuracy
print('Training Score',m1.score(x_train,y_train))
print('Testing Score',m1.score(x_test,y_test))
```

```python
ypred_m1 = m1.predict(x_test)
print(ypred_m1)
```

```python
gen_metrics(y_test,ypred_m1)
```

Validation

```python
pre0 = 4834/(4834+718)
pre1 = 1251/(1251+738)
rec0 = 4834/(4834+738)
rec1 = 1251/(1251+718)
print('Pre0',pre0)
print('Pre1',pre1)
print('Rec0',rec0)
print('Rec1',rec1)
```

```python
f1s0 = 2*pre0*rec0/(pre0+rec0)
f1s1 = 2*pre1*rec1/(pre1+rec1)
print('F1_Score0',f1s0)
print('F1_Score1',f1s1)
```

```python
acc = (4834+1251)/(4834+738+718+1251)
print('Accuracy',acc)
```

Percentage of misclasification

```python
# Misclassification Rate = incorrect predictions / total predictions
# Misclassification Rate = (false positive + false negative) / (total predictions)
MR = (718+738)/(4834+738+718+1251)
print('Percentage of Misclassification : ',MR*100, '%')
```
